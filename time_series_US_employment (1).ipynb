```python
#This dataset contains the U.S. unemployment rate by industry from 2000-2010
unemployment = pd.read_csv("https://assets.datacamp.com/production/course_5226/datasets/ch5_employment.csv")
unemployment.head()
```

```python
#Plot the time series in the data. Use subplots to make the line charts legible.
unemployment.plot(subplots=True,
                 layout=(4,4),
                 figsize=(20,16),
                 sharex=True,
                 sharey=False)
```

```python
# Plot monthly trends
# Extract the month from the index of unemployment
index_month = unemployment.index.month

# Calculate the mean unemployment rate for each month
unemployment_by_month = unemployment.groupby(index_month).mean()

# Plot the mean unemployment rate for each month
ax = unemployment_by_month.plot(fontsize=8, linewidth=1)

# Set axis labels and legend
ax.set_xlabel('Month', fontsize=10)
ax.set_ylabel('Mean Unemployment Rate', fontsize=10)
ax.legend(bbox_to_anchor=(1, 1.1), fontsize=10)
```

```python
# Plot yearly trends
# Extract of the year in each date index of jobs data frame
index_year = unemployment.index.year

# Calculate the mean unemployment rate for each year
unemployment_by_year = unemployment.groupby(index_year).mean()

# Plot the mean unemployment rate for each year
ax = unemployment_by_year.plot(fontsize=6, linewidth=1)

# Set axis labels and legend
ax.set_xlabel('Year', fontsize=10)
ax.set_ylabel('Mean Unemployment Rate', fontsize=10)
ax.legend(bbox_to_anchor=(1, 1.1), fontsize=10)
```

```python
# Decompose the time series data

# Import statsmodel library
import statsmodels.api as sm

# Create a dictionary to store results of each time series decomposition 
unemployment_dict = {}

# Pull name of each industry in the data frame
industry_names = unemployment.columns

# Run time series decomposition on each time series in the data frame
for ts in industry_names:
    ts_decomposition = sm.tsa.seasonal_decompose(unemployment[ts])
    unemployment_dict[ts] = ts_decomposition
```

```python
#Visualize seasonality of multiple time series. 
#For more on seasonality, view this post: https://machinelearningmastery.com/time-series-seasonality-with-python/

# Create a dictionary to store results of seasonality 
unemployment_seasonal = {}

# Extract the seasonal values for the decomposition of each time series
for ts in industry_names:
    unemployment_seasonal[ts] = unemployment_dict[ts].seasonal
    
# Create a DataFrame from the jobs_seasonal dictionnary
seasonality_df = pd.DataFrame.from_dict(unemployment_seasonal)

# Remove the label for the index
seasonality_df.index.name = None

# Create a faceted plot of the seasonality_df DataFrame
seasonality_df.plot(subplots=True,
                   layout=(4,4),
                   sharey=False,
                   fontsize=6,
                   linewidth=0.3,
                   legend=False)
```

```python
#Show correlations between multiple time series

#import seaborn library
import seaborn as sns

# Display correlation matrix of the seasonality_df DataFrame using Spearman method of correlation
seasonality_corr = seasonality_df.corr(method='spearman')

# Clustermap of the seasonality_corr correlation matrix
fig = sns.clustermap(seasonality_corr, annot=True, annot_kws={"size": 4}, linewidths=.4, figsize=(16, 8))
plt.setp(fig.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
plt.setp(fig.ax_heatmap.xaxis.get_majorticklabels(), rotation=90)
```
